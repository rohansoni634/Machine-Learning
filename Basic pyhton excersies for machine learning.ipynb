<h3> Q2: Select a number randomly with probability proportional to its magnitude from the given array of n elements</h3>

<pre>
Ex 1: A = [0 5 27 6 13 28 100 45 10 79]
let f(x) denote the number of times x getting selected in 100 experiments.
f(100) > f(79) > f(45) > f(28) > f(27) > f(13) > f(10) > f(6) > f(5) > f(0)

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples
# you can free to change all these codes/structure
from collections import Counter
from bisect import bisect
from random import uniform 

A = [0,5,27,6,13,28,100,45,10,79]
def pick_a_number_from_list(A):
    sum=0
    d_dash=[]
    
    #step 1 : getting sum of all elements in the list
    for i in range(0,len(A)):
        sum+=A[i]
    #print(sum)
    
    #step 2: normalising the value using the sum
    for i in range(0,len(A)):
        d_dash.append(A[i]/sum)
    #print('D_dash : ',d_dash)
    
    #step 3: finding commulative sum
    d_doulbe_dash=[]
    c_sum=0
    for x in d_dash:
        c_sum+=x
        d_doulbe_dash.append(c_sum)
    #print('d_doulbe_dash : ',d_doulbe_dash) 
    
    #getting a random number between 0 and 1
    r=uniform(0.0,1.0)   
    #print('R : ',r)

    idx=bisect(d_doulbe_dash,r*d_doulbe_dash[-1])
    '''d_doulbe_dash[-1] return the max value in d_doulbe_dash.multiplyong it with r gives a value netween 0 and 
    d_doulbe_dash[-1].Finally bisect returns the index of the value in d_doulbe_dash which is closed to it,and thus
    we can find the value from A prsent at that index'''
    result=A[idx]
    return result

def sampling_based_on_magnitued():
    out_list=[]
    for i in range(100):#repeating the experiment 100 times
        out_list.append(pick_a_number_from_list(A))
    #print(out_list) 
    Result=Counter(out_list)#counting the value and it occurence
    print(Result)
    
sampling_based_on_magnitued()  
```

<h3> Q5: Find the closest points</h3>

Consider you are given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P

<br>Cosine distance between two points (x,y) and (p,q) is defined as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math 

dis_lst=[]

def closest_points_to_p(S,P):
    '''This function returns 5 closeset points form a list of tuple containing n data points '''
    for i in range(len(S)):
        d=((S[i][0]*P[0])+(S[i][1]*P[1]))/(math.sqrt(((S[i][0]**2)+(S[i][1]**2))*((P[0]**2)+(P[1]**2))))
        #taking cos inverse and appeding the result of each iteration into an empty list
        dis_lst.append(math.acos(d))
    mapped=zip(S,dis_lst) 
    mapped=list(mapped)
    #sorting the list 
    mapped.sort(key=lambda x:x[1])
    #unzipping the list to seperate points and angles obtained after applying cos inverse function
    points,distance=zip(*mapped)
    return points[:5]


S=[(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P=(3,-4)
# calling the function and storing the result in a variable
closest_points=closest_points_to_p(S,P)
print(closest_points)
        
       
```

```python
#ref: https://stackoverflow.com/questions/57160252/find-conditional-probabilities-using-python
from fractions import Fraction
A=[['F1','S1'],['F2','S2'],['F3','S3'],['F1','S2'],['F2','S3'],['F3','S2'],['F2','S1'],['F4','S1'],['F4','S3'],['F5','S1']]
def compute_conditional_probabilites(F,S):
    num=0
    den=0
    for i in range(len(A)):
        if(A[i][1]==S):
            den=den+1
            if(A[i][0]==F):
                num=num+1
    
    print('P(F={}|S=={})='.format(F,S), Fraction(num,den))
        
for k in ['F1', 'F2', 'F3', 'F4', 'F5']:
    for m in ['S1', 'S2', 'S3']:
        compute_conditional_probabilites(k,m)

```

<h3> Q10: Error Function</h3>

You will be given a list of lists, each sublist will be of length 2 i.e. [[x,y],[p,q],[l,m]..[r,s]]
consider its like a martrix of n rows and two columns

a. the first column Y will contain interger values <br>
b. the second column $Y_{score}$ will be having float values <br>
Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.44982
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python

import math as m

def compute_log_loss(A):
    n=len(A)
    loss=0
    for i in range(0,n):
        loss+=(A[i][0]*(m.log10(A[i][1])))+((1-A[i][0])*(m.log10(1-A[i][1])))
    loss=-(loss/n)   
    return loss

A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
loss = compute_log_loss(A)
print(loss)
```
